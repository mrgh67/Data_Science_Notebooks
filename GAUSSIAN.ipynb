```python
mu = 0.0
```

```python
sigma = 1.0
```

```python
def norm(x, mu, sigma):
    sigma_squared = sigma * sigma
    temp = 1.0/np.sqrt(2.0 * np.pi * sigma_squared)
    temp1 = np.exp(-(1.0/2.0*sigma_squared) * (x - mu) * (x - mu))
    return temp * temp1
```

```python
norm(0, mu, sigma)
```

```python
fig1 = plt.figure(1, figsize=(8,5))
ax1 = plt.subplot(111)
plt.xlim(-5,5)
plt.ylim(0,0.5)

x = np.linspace(-5, 5, 100)

ax1.plot(x, norm(x, mu, sigma),'r-', lw=1, alpha=0.6, label='normal pdf')
ax1.set_title("Normal Distribution mu = %s, sigma = %s"%(str(mu), str(sigma)))
```

```python
def norm_D(X, MU, SIGMA, D):
    temp1 = 1.0/np.power(2.0 * np.pi, float(D)/2.)
    #print "temp1 is ", temp1
    
    temp2 = 1.0/np.sqrt(np.linalg.det(SIGMA))
    #print "temp2 is ", temp2
    
    Xm = np.matrix(X)
    #print "shape Xm is ", Xm.shape
    
    MUm = np.matrix(MU)
    #print "shape MUm is ", MUm.shape
    
    temp3 = np.matrix(Xm-MUm)
    #print "temp3 is ", temp3
    #print "temp3 shape is ", temp3.shape
    
    temp4 = np.dot(temp3, np.linalg.inv(SIGMA))
    #print "temp4 is", temp4
    
    temp5 = np.dot(temp4, np.transpose(temp3))
    #print "temp5 is ", temp5
    #print "temp5 shape is ", temp5.shape
    #print np.matrix.item(temp5)
    
    temp6 = -0.5 * np.matrix.item(temp5)
    
    return temp1 * temp2 * np.exp(temp6)
```

##$\bigtriangleup^{2} = (x - \mu)^{T} \Sigma^{-1} (x - \mu)$

##$\bigtriangleup \text{ is called the Mahalonobis Distance}$

##$\text{If } \Sigma \text{ is the identity matrix then this reduces to the Euclidean Distance}$

##$\text{The matrix } \Sigma \text{ is symmetric, i.e. a square matrix that is equal to its transpose}$

##$\text{The Eigenveector equation for the covariance matrix is  } \Sigma u_{i} = \lambda_{i} u_{i} \text{ where } i = 1 ,\dotsc, D$

##$\Sigma \text{ is real and symmetric, and therefore it's eigenvalues will be real, and its eigenvectors can be chosen}$

##$\text{to form an orthonormal set}$

##$u^{T}_{i}u_{j} = I_{ij} \text{ where } I_{ij} = 1, \text{ if }i=j, 0 \text{ otherwise}$

##$\Sigma \text{ may be expressed as an expansion in terms of its eigenvector in the form: }\Sigma=\sum^{D}_{i=1}\lambda_{i}u_{i}u^{T}_{i}$

##$\text{and similiarly :}\Sigma^{-1}=\sum^{D}_{i=1}\frac{1}{\lambda_{i}}u_{i}u^{T}_{i}$

##$\text{If we let }y_{i} = u^{T}_{i}(x-\mu) \text{ then } \bigtriangleup^{2} = \sum^{D}_{i=1} \frac{y^{2}_{i}}{\lambda_{i}}$

##$\text{We interpret }y_{i} \text{ as a new coordinate system defined by the orthonormal vectors } u_{i} \text{, that are shifted}$ 

##$\text{and rotated with respect to the original } x_{i} \text{ coordinates}$

##$\text{Forming the vector } y = (y_{1},\dotsc.y_{D})^{T} \text{, we have : } y = U(x-\mu)$

##$\text{ where U is a matrix whose rows are given by } u^{T}_{i}$

##$U \text{ is an orthogonal matrix, satisfying } UU^{T} = I \text{, and also } U^{T}U = I $
