```python
import pandas as pd
import numpy as np
from scipy.stats import beta, norm, binom
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Math
```

###ctr = click through rate
###views = page views

```python
def percent_better(a, b, a1, b1, size):
    ra = beta.rvs(a, b, size=(size))
    rb = beta.rvs(a1, b1, size=(size))
    return sum(ra >= rb) / float(size)
```

```python
def demonstrate(a_views, a_ctr, b_views, b_ctr, size):
    
    fig, ax = plt.subplots(1, 1)
    fig.figsize=(15, 8)
    
    a = a_views * a_ctr;
    b = a_views * (1. - a_ctr)
    upper = beta.ppf(1.0 - 0.0001, a, b)
    
    print "a_views = ", a_views
    print "a_ctr = ", a_ctr
    print "a_views * a_ctr = ", a
    print "a_views * (1. - a_ctr) = ", b
    print "upper = ", upper
    
    xs = np.linspace(0, upper, 2000)
    #print "xs is ", xs
    
    ys = [beta.pdf(x, a, b) for x in xs]
    #print "ys is ", ys[0:10]
    print "sum ys = ", sum(ys)
    print "len ys = ", len(ys)
    
    ys /= sum(ys) / float(len(ys))
    
    #print "ys is ", ys[0:10]
    ax.plot(xs, ys)
    
    ax.hold(True)
    
    a2 = b_views * b_ctr;
    b2 = b_views * (1. - b_ctr)
    upper2 = beta.ppf(1.0 - 0.0001, a2, b2)
    print "b_views = ", b_views
    print "b_ctr = ", b_ctr
    print "b_views * b_ctr = ", a2
    print "b_views * (1. - b_ctr) = ", b2
    print "upper2 = ", upper2
    
    pb = percent_better(a, b, a2, b2, size)
    print "pb is ", pb
    rpb = percent_better(a2, b2, a, b, size)
    print "rpb is ", rpb
    
    if pb > 0.5:
        winner = 'A'
    else:
        winner = 'B'
        pb = 1.0 - pb
    
    xs2 = np.linspace(0, upper2, 2000)
    ys2 = [beta.pdf(x, a2, b2) for x in xs2]
    ys2 /= sum(ys2) / float(len(ys2))
    ax.plot(xs2, ys2)

    #ax.legend(['Headline A (views: %s, CTR: %0.2f)'% (a_views, a_ctr),'Headline B (views: %s, CTR: %0.2f)' % (b_views, b_ctr)]);
    ax.set_title('Headline %s is better in %.2f%% of the cases' % (winner, pb*100.0));
    ax.set_xlabel('CTR')
```

```python
A_views = 100
A_ctr = 0.04969
B_views = 200
B_ctr = 0.13287
demonstrate(A_views, A_ctr, B_views, B_ctr, size=1000000)
```

#Anscombe threshold

Anscombe provides a formula to determine the stopping point of an experiment. The experiment should be terminated when the following condition is true.

Where y is the difference between results of A and B, k is the expected number of future users who will be exposed to a result, and n is the number of users who are exposed to the test so far. And  Phi-inverse is the quantile function of the standard normal.

```python
Math(r'|c_a - c_b| > -\Phi^{-1} \left( \frac{n}{k+2n} \right) \sqrt(n)')
```

```python
fig, ax = plt.subplots(1, 1)
fig.figsize=(45, 45)


Views_per_Minute = 150
Duration = 5
Future = 20

A_ctr = 0.05
B_ctr = 0.049

n = np.arange(Duration * Views_per_Minute)
n = n + 1
#print n
k = Future * Views_per_Minute
Ansc = -norm.ppf(n/(k+(2.0*n))) * np.sqrt(n)
#print Ansc

A_clicks = binom.rvs(n, 0.051)
#print A_ctr * n
#print A_clicks
B_clicks = binom.rvs(n, 0.041)
#B_ctr * n
#print B_clicks
#print np.abs(A_clicks-B_clicks)

x = np.arange(Views_per_Minute * Duration)
ax.plot(x, Ansc)
ax.hold(True)
ax.plot(x, -Ansc)
ax.hold(True)
ax.plot(x, A_clicks-B_clicks)
```
